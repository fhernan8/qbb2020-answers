# Implement a script that finds matching k-mers between a single query sequence (droyak2.seq) and a database of targets (subset.fa) 
The matcher should take three arguments: kmer_matcher.py <target.fa> <query.fa> <k>
Where target.fa is the database, potentially multiple sequences, query.fa is the sequence to align (assume just one sequnce), and k is the k-mer size (an integer).

The script should find k-mer matches and for each write: target_sequence_name    target_start    query_start k-mer
For submission: Run the program for k=11 and submit the first 1000 lines, along with your script.
    

```python
from fasta_iterator_class import FASTAReader
Reader = FASTAReader(open('subset.fa'))
target_kmers = {} 
k = 11
for seq_id, sequence in Reader: 
    for i in range(len(sequence)-k+1):
        kmer = sequence[i: i+k]
        if kmer in target_kmers:
            target_kmers[kmer].append((seq_id, i))
        else:
            target_kmers[kmer] = [seq_id, i]

print_counter = 0
query_reader = FASTAReader(open('droYak2_seq.fa'))
for seq_id_q, sequence_q in query_reader: 
    for j in range(len(sequence_q)-k+1):
        kmer_q = sequence_q[j:j+k]
        if kmer_q in target_kmers:
            seq_id_position_info = target_kmers[kmer_q]
            for n in range (len(seq_id_position_info)):
                #seq_id, 
                i = seq_id_position_info[n]
                print_counter += 1
                if print_counter <= 1000:
                    print(seq_id, i, kmer_q, j)
                else:
                    quit()
                
```
