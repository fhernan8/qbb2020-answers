Load Modules

```python

bed_columns = ["chr", "start", "end"]
df_k4 = pd.read_table("~/qbb2020-answers/K4me3.bed", names=bed_columns)
df_k4

bed_columns = ["chr", "start", "end"]
df_k9 = pd.read_table("~/qbb2020-answers/K9me3.bed", names=bed_columns)
df_k9

bed_columns = ["chr", "start", "end"]
df_k27 = pd.read_table("~/qbb2020-answers/K27me3.bed", names=bed_columns)
df_k27
```

Get Crhomosomal Distribution and Sort It

```python
chr_k4 = df_k4.loc[:,"chr"]
chr_k4 = chr_k4.value_counts()
chr_k4 = chr_k4.sort_index()
chr_k4

chr_k9 = df_k9.loc[:,"chr"]
chr_k9 = chr_k9.value_counts()
chr_k9 = chr_k9.sort_index()
chr_k9

chr_k27 = df_k27.loc[:,"chr"]
chr_k27 = chr_k27.value_counts()
chr_k27 = chr_k27.sort_index()
chr_k27
```

Plot sorted chromosomal distribution 

```python
fig, ax = plt.subplots()
ax.bar(chr_k4.index, chr_k4)
plt.xticks(rotation=45, ha='right') 
ax.set_xlabel("chromsomes")
ax.set_ylabel("Number of chromosomes")
ax.set_title("Distribution of chromosomes in K4")
plt.show()
```

```python
fig, ax = plt.subplots (nrows=3) 
fig.set_figheight(15) 
ax[0].bar(chr_k4.index, chr_k4)
ax[1].bar(chr_k9.index, chr_k9)
ax[2].bar(chr_k27.index, chr_k27)
ax[0].set_xticklabels(chr_k4.index, rotation=45, ha="right")
ax[1].set_xticklabels(chr_k9.index, rotation=45, ha="right")
ax[2].set_xticklabels(chr_k27.index, rotation=45, ha="right")
ax[0].set_xlabels("chromosomes")
ax[1].set_xlabels("chromosomes")
ax[2].set_xlabels("chromosomes")
ax[0].set_ylabels("Number of chromosomes")
ax[1].set_ylabels("Number of chromosomes")
ax[2].set_ylabels("Number of chromosomes")
ax[0].set_title ("Distribution of Chromosomes in K4")
ax[1].set_title ("Distribution of Chromosomes in K9")
ax[2].set_title ("Distribution of Chromosomes in K27")
fig,tight_layout(pad=5.0)
```

Adding widths to Dataframe

```python
df_k4.loc[:,"width"] = df_k4.loc[:,"end"] - df_k4.loc[:,"start"]
df_k4

df_k9.loc[:,"width"] = df_k9.loc[:,"end"] - df_k9.loc[:,"start"]
df_k9

df_k27.loc[:,"width"] = df_k27.loc[:,"end"] - df_k27.loc[:,"start"]
df_k27
```

```python
fig, ax = plt.subplots (nrows=3)
fig.set_figheight(15)
ax[0].hist(df_k4.loc[:, "width"])
ax[1].hist(df_k9.loc[:, "width"])
ax[2].hist(df_k27.loc[:, "width"])
ax[0].tick_params(rotation=45)
ax[2].tick_params(rotation=45)
ax[0].set_xlabel("Width")
ax[1].set_xlabel("Width")
ax[2].set_xlabel("Width")
ax[0].set_ylabel("Count")
ax[1].set_ylabel("Count")
ax[2].set_ylabel("Count")
ax[0].set_title("Width distribution in K4")
ax[1].set_title("Width distribution in K9")
ax[2].set_title("Width distribution in K27")
fig.tight_layout(pad=5.0)
```
